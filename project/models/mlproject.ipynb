```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Assuming df is already loaded as your dataset
# Preprocess the data
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Feature Engineering
df['precipitation_rolling_mean'] = df['precipitation'].rolling(window=7).mean()
df['temp_min_rolling_mean'] = df['temp_min'].rolling(window=7).mean()
df['wind_rolling_mean'] = df['wind'].rolling(window=7).mean()

# Fill NaN values resulting from rolling mean
df.fillna(method='bfill', inplace=True)

# Selecting features
features = df[['precipitation', 'temp_min', 'wind', 'precipitation_rolling_mean', 'temp_min_rolling_mean', 'wind_rolling_mean']]
target = df['temp_max']

# Scaling the features
feature_scaler = MinMaxScaler(feature_range=(0, 1))
scaled_features = feature_scaler.fit_transform(features)

target_scaler = MinMaxScaler(feature_range=(0, 1))
scaled_target = target_scaler.fit_transform(target.values.reshape(-1, 1))

# Create sequences
def create_sequences(data, target, sequence_length=50):
    sequences = []
    for i in range(len(data) - sequence_length):
        sequences.append((data[i:i+sequence_length], target[i+sequence_length]))
    return sequences

sequence_length = 50
sequences = create_sequences(scaled_features, scaled_target, sequence_length)

X, y = zip(*sequences)
X = np.array(X)
y = np.array(y)

# Split into training and testing sets
split_ratio = 0.8
split_index = int(len(X) * split_ratio)

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(sequence_length, X_train.shape[2])))
model.add(Dropout(0.3))
model.add(LSTM(128, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(1))

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Callbacks for early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test),
                    callbacks=[early_stopping, reduce_lr])

# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions and actual values
predictions = target_scaler.inverse_transform(predictions)
y_test = target_scaler.inverse_transform(y_test)

# Calculate accuracy metrics
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot the results
plt.figure(figsize=(14, 5))
plt.plot(df.index[split_index+sequence_length:], y_test, color='blue', label='Actual Temperature')
plt.plot(df.index[split_index+sequence_length:], predictions, color='red', label='Predicted Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (max)')
plt.title('Temperature Prediction using LSTM')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.model_selection import train_test_split

# Assuming df is already loaded as your dataset
# Preprocess the data
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Feature Engineering (same as before)
df['precipitation_rolling_mean'] = df['precipitation'].rolling(window=7).mean()
df['temp_min_rolling_mean'] = df['temp_min'].rolling(window=7).mean()
df['wind_rolling_mean'] = df['wind'].rolling(window=7).mean()

# Fill NaN values resulting from rolling mean
df.fillna(method='bfill', inplace=True)

# Selecting features and target
target = df['temp_max']

# Split into training and testing sets
train_target, test_target = train_test_split(target, test_size=0.2, shuffle=False)

# Build and fit the ETS model
model = ExponentialSmoothing(train_target, trend='add', seasonal='add', seasonal_periods=12)
model_fit = model.fit()

# Forecast future values
forecast = model_fit.forecast(steps=len(test_target))

# Calculate accuracy metrics
mae = mean_absolute_error(test_target, forecast)
mse = mean_squared_error(test_target, forecast)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot the results
plt.figure(figsize=(14, 5))
plt.plot(test_target.index, test_target, color='blue', label='Actual Temperature')
plt.plot(test_target.index, forecast, color='red', label='Predicted Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (max)')
plt.title('Temperature Prediction using ETS')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import joblib

# Load the data
df = pd.read_csv('D:\PROJECT\ML Project\seattle-weather (1).csv', index_col='date', parse_dates=['date'])

# Select only the temp_max column
temp_max = df['temp_max']

# Scale the data
scaler = MinMaxScaler()
scaled_temp_max = scaler.fit_transform(temp_max.values.reshape(-1, 1))

# Save the scaler
joblib.dump(scaler, 'scaler.joblib')

# Create sequences for LSTM
sequence_length = 50
sequences = []
for i in range(len(scaled_temp_max) - sequence_length):
    sequences.append(scaled_temp_max[i:i+sequence_length])
sequences = np.array(sequences)

# Split into training and testing sets
split_ratio = 0.8
split_index = int(len(sequences) * split_ratio)

X_train, X_test = sequences[:split_index], sequences[split_index:]
y_train, y_test = scaled_temp_max[sequence_length:split_index+sequence_length], scaled_temp_max[split_index+sequence_length:]

# Reshape the data for LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build and compile the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(sequence_length, 1)))
model.add(Dropout(0.3))
model.add(LSTM(128, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Callbacks for early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

# Train the model
history = model.fit(X_train, y_train, epochs=40, batch_size=16, validation_data=(X_test, y_test),
                    callbacks=[early_stopping, reduce_lr])

# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions and actual values
scaler = joblib.load('scaler.joblib')
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test)

# Calculate accuracy metrics
mae = np.mean(np.abs(predictions - y_test))
mse = np.mean((predictions - y_test) ** 2)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot the results
plt.figure(figsize=(14, 5))
plt.plot(y_test, label='Actual Temperature')
plt.plot(predictions, label='Predicted Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (max)')
plt.title('Temperature Prediction using LSTM')
plt.legend()
plt.show()

# Save the trained model
model.save('lstm_temp_max_model.h5')

# Predict future temperatures
future_days = 30
future_input = scaled_temp_max[-sequence_length:]
future_input = future_input.reshape((1, sequence_length, 1))
future_predictions = model.predict(future_input)
future_predictions = scaler.inverse_transform(future_predictions)

print(f'Future temperature predictions for the next {future_days} days:')
print(future_predictions)
```
